```python
#module import

import pandas as pd
```

```python
#서울시 cctv, 과속방지카메라, 인구 밀도, 사고수, 어린이 시설 비율, 어린이 보호구역 파일 불러오기

cctv = pd.read_excel('seoul_cctv.xlsx', encoding='utf-8')
speed_camera = pd.read_excel('서울시 과속방지카메라.xlsx', encoding='utf-8')
pop_ratio = pd.read_excel('서울시_인구밀도.xlsx', encoding='utf-8')
accident = pd.read_csv('사고통계.csv')
facility = pd.read_excel('서울시 어린이 시설 비율.xlsx', encoding='utf-8')
protect = pd.read_excel('서울_보호구역추가.xlsx', encoding='utf-8')
```

```python
#cctv.head()
```

```python
#camera.head()
```

```python
#필요 없는 열 삭제

accident = accident.iloc[:,1:].reset_index(drop=True)
```

```python
#accident.head()
```

```python
#facility.head()
```

```python
#데이터 프레임 합치기

seoul_final = pd.concat([protect,accident[['사고통계']],cctv[['cctv_total','cctv_kids']],speed_camera[['num_camera']],facility[['facility_ratio']]],axis=1)
```

```python
#seoul_final.head()
```

```python
#구별 값을 인구 밀도로 나눈 비율 계산 

seoul_final['accident_ratio'] = seoul_final['사고통계'] / seoul_final['pop_dens']
seoul_final['cctv_ratio'] = seoul_final['cctv_total'] / seoul_final['pop_dens']
seoul_final['cctv_kids_ratio'] = seoul_final['cctv_kids'] / seoul_final['pop_dens']
seoul_final['speedcamera_ratio'] = seoul_final['num_camera'] / seoul_final['pop_dens']

seoul_final
```

```python
seoul_final.to_excel("서울시 최종 비율.xlsx")
```

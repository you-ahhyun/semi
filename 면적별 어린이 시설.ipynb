```python
#import module

import pandas as pd
```

```python
#서울시 어린이집, 유치원, 초등학교, 면적, 인구밀도 파일 불러오기

num_ele = pd.read_excel('num_elementary.xlsx', encoding='utf-8')
num_nursey = pd.read_excel('num_nursery.xlsx', encoding='utf-8')
num_kindergarten = pd.read_excel('num_kindergarten.xlsx', encoding='utf-8')
total_area = pd.read_excel('toal_area.xlsx', encoding='utf-8')
pop_ratio = pd.read_excel('서울시_인구밀도.xlsx', encoding='utf-8')
```

```python
#서울시 어린이집, 유치원, 초등학교 개수 더하기

sum_facility = pd.concat([num_ele,num_nursey[['계 (개소)']],num_kindergarten[['개수']]], axis=1)
sum_facility['total'] = sum_facility.sum(axis=1)
```

```python
#서울시 총 어린이 시설 개수와 인구밀도 데이터프레임 합치기

facility = pd.concat([sum_facility,pop_ratio[['인구밀도']]], axis=1)
```

```python
#서울시 인구밀도 대비 어린이 시설 개수

facility['pop_ratio'] = pop_ratio['인구밀도']
facility['facility_ratio'] = facility['total'] / pop_ratio['인구밀도']
```

```python
#데이터프레임 열 이름 바꾸기

del facility["인구밀도"]
facility.columns = ["region", "elementary","nursey", 'kindergarten', "total", "pop_ratio","per_facility"]
```
